```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
rainbow = pickle.load(open("models/rainbow_simple.pkl", "rb"))
simple1 = pickle.load(open("models/simple_simple.pkl", "rb"))
simple2 = pickle.load(open("models/simple_simple2.pkl", "rb"))
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
returns_rb = np.asarray(rainbow[0])
returns_s1 = np.asarray(simple1[0])
returns_s2 = np.asarray(simple2[0])

fig=plt.figure(figsize=(6,3.8))
plt.plot(running_mean(returns_rb, 100), label="Rainbow")
plt.plot(running_mean(returns_s1, 100), label="DQN")
plt.plot(running_mean(returns_s2, 100), label="DQN-edited")
plt.legend()
```

```python
losses_rb = np.asarray(rainbow[1])
losses_s1 = np.asarray(simple1[1])
losses_s2 = np.mean(np.asarray(simple2[1]).reshape(-1, 32), axis=1)

fig=plt.figure(figsize=(6,3.8))
plt.plot(losses_rb, label="Rainbow")
plt.plot(losses_s1, label="DQN")
plt.plot(losses_s2, label="DQN-edited")
plt.legend()
```

```python
# Data to plot
labels = ['Tournament', 'Baseline', 'Strong', "Self-Play"]
values1 = [0.67, 0.66, 0.78, 0.78]
values2 = [0.37, 0.31, 0.43, 0.43]

# Create a bar chart
fig, ax = plt.subplots()
ax.bar(labels, values1, alpha=0.5, label='Winrate against strong', color='tab:blue')
ax.bar(labels, values2, label='Winrate against weak', color='tab:blue')

# Add title and labels
ax.set_title('Winrates by agent')
ax.set_ylabel('Winrate [%]')
plt.legend()

# Show the plot
plt.show()
```
